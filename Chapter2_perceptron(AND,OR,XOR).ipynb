```python
# simple perceptron example
# 여러 개의 입력 신호에 가중치를 곱해 더한 뒤, 그 합이 임계값을 넘으면 1을 출력하는 가장 기본적인 신경망 단위

def AND(x1,x2):
  w1, w2 ,theta = 0.5,0.5,0.7
  tmp = x1*w1 + x2*w2
  if tmp > theta:
    return 1
  else:
    return 0


print("AND(1,0):", AND(1,0))
print("AND(1,1):", AND(1,1))

```

```python
# perceptron with weight

import numpy as np

def AND(x1,x2):
  x = np.array([x1,x2])
  w = np.array([0.5,0.5])
  bias = -0.6

  tmp = np.sum(w*x) + bias
  if tmp > 0:
    return 1
  else:
    return 0


print("AND(1,0)" , AND(1,0))
print("AND(1,1)" , AND(1,1))
print()

def NAND(x1,x2):
  x = np.array([x1,x2])
  w = np.array([-0.5,-0.5])
  bias = 0.6
  tmp = np.sum(x*w) + bias

  if tmp > 0:
    return 1
  else:
    return 0

print("NAND(0,0)", NAND(0,0))
print("NAND(0,1)", NAND(0,1))
print("NAND(1,0)", NAND(1,0))
print('NAND(1,1)', NAND(1,1))
print()

def OR(x1,x2):
  x = np.array([x1,x2])
  w = np.array([0.5,0.5])
  bias = -0.1
  tmp = np.sum(w*x) + bias

  if tmp > 0:
    return 1
  else:
    return 0


print('OR(1,0)', OR(1,0))
```

```python
#multi perceptron can solve XOR problem

def XOR(x1,x2):
  s1 = NAND(x1,x2)
  s2 = OR(x1,x2)
  y = AND(s1,s2)
  return y

print('XOR(0,0)', XOR(0,0))
print('XOR(0,1)', XOR(0,1))
print('XOR(1,0)', XOR(1,0))
print('XOR(1,1)', XOR(1,1))


```
