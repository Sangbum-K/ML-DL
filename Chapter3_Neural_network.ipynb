```python
# Neural network: 기계학습과 인지과학에서 생물학의 신경망에서 영감을 얻은 알고리즘
# Activation function: 입력신호의 총합을 출력신호로 변환해주는 함수
```

```python
# Step function: 임곗값을 경계로 출력을 바꿔주는 활성화 함수

import numpy as np

def step_function(x):
  y = x > 0
  return y.astype(int)

x = np.array([-1,2,3])

print(step_function(x))
print(type(step_function(x)))
```

```python
def step_function(x):
  return np.array(x > 0 ,dtype = int)

x = np.array([-1,2,3])

print(step_function(x))
print(type(step_function(x)))
```

```python
import matplotlib.pylab as plt

x = np.arange(-5,5,0.01)
y = step_function(x)
plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```

```python
# Sigmoid function: 입력값을 0과 1 사이의 실수로 변환하는 비선형 활성화 함수

def sigmoid(x):
  return 1/(1+np.exp(-x))

x = np.arange(-5,5,0.01)
y = sigmoid(x)

plt.plot(x,y)
plt.ylim(-0.1,1.1)
plt.show()
```

```python
# ReLU: Rectified Linear Unit

def ReLU(x):
  return np.maximum(0,x) #maximum 함수는 요소별 비교

x = np.arange(-5,5,0.01)
y = ReLU(x)

plt.plot(x,y)
plt.ylim(-0.1,5)
plt.show()
```
